```python
#The Model
#=======================================================
def eq(par,initial_cond,start_t,end_t,incr):
    #-time-grid-----------------------------------
    t  = np.linspace(start_t, end_t,incr)
    #differential-eq-system---------------------
    def funct(y,t):
        OTU1i = y[0]
        OTU2i = y[1]
        #OTU3i = y[2]
        #OTU4i = y[3]
        #OTU5i = y[4]
        #OTU6i = y[5]
        #OTU7i = y[6] 

        #Parameters
        #a1,a2,a3,a4,a5,a6,a7,b12,b13,b14,b15,b16,b17,b21,b23,b24,b25,b26,b27,b31,b32,b34,b35,b36,b36,b37,b41,b42,b43,b45,b46,b47,b51,b52,b53,b54,b56,b57,b61,b62,b63,b64,b65,b67,b71,b72,b73,b74,b75,b76 = par 
        a1,a2,b12,b21 = par 


        OTU1 = a1*OTU1i*(1-(OTU1i/1) + OTU1i*(b12*OTU2i))#+b13*OTU3i+b14*OTU4i+b15*OTU5i+b16*OTU6i+b17*OTU7i))
        OTU2 = a2*OTU2i*(1-(OTU2i/1) + OTU2i*(b21*OTU1i))#+b23*OTU3i+b24*OTU4i+b25*OTU5i+b26*OTU6i+b27*OTU7i))
       # OTU3 = a3*OTU3i*(1-(OTU3i/1) + OTU3i*(b31*OTU1i+b32*OTU2i+b34*OTU4i+b35*OTU5i+b36*OTU6i+b37*OTU7i))
       # OTU4 = a4*OTU4i*(1-(OTU4i/1) + OTU4i*(b41*OTU1i+b42*OTU2i+b43*OTU3i+b45*OTU5i+b46*OTU6i+b47*OTU7i))
       # OTU5 = a5*OTU5i*(1-(OTU5i/1) + OTU5i*(b51*OTU1i+b52*OTU2i+b53*OTU3i+b54*OTU4i+b56*OTU6i+b57*OTU7i))
       # OTU6 = a6*OTU6i*(1-(OTU6i/1) + OTU6i*(b61*OTU1i+b62*OTU2i+b63*OTU3i+b64*OTU4i+b65*OTU5i+b67*OTU7i))
       # OTU7 = a7*OTU7i*(1-(OTU7i/1) + OTU7i*(b71*OTU1i+b72*OTU2i+b73*OTU3i+b74*OTU4i+b75*OTU5i+b76*OTU6i))
        
        
        return [OTU1,OTU2]#,OTU3,OTU4,OTU5,OTU6,OTU7]
        
        
   #integrate------------------------------------
    ds = integrate.odeint(funct,initial_cond,t, mxstep=5000)
    return (ds[:,0],ds[:,1],t)#,ds[:,2],ds[:,3],ds[:,4],ds[:,5],ds[:,6],t)
#=======================================================
        
        
        
        
        
        
```

```python
#2.Set up Info for Model System
#===================================================
# model parameters
#----------------------------------------------------
a1=1
a2=1
#a3=1
#a4=1
#a5=1
#a6=1
#a7=1

b12=1
#b13=1
#b14=1
#b15=1
#b16=1
#b17=1

b21=1
#b23=1
#b24=1
#b25=1
#b26=1
#b27=1

#b31=1
#b32=1
#b34=1
#b35=1
#b36=1
#b37=1

#b41=1
#b42=1
#b43=1
#b45=1
#b46=1
#b47=1

#b51=1
#b52=1
#b53=1
#b54=1
#b56=1
#b57=1

#b61=1
#b62=1
#b63=1
#b64=1
#b65=1
#b67=1

#b71=1
#b72=1
#b73=1
#b74=1
#b75=1
#b76=1
#rates=(a1,a2,a3,a4,a5,a6,a7,b12,b13,b14,b15,b16,b17,b21,b23,b24,b25,b26,b27,b31,b32,b34,b35,b36,b36,b37,b41,b42,b43,b45,b46,b47,b51,b52,b53,b54,b56,b57,b61,b62,b63,b64,b65,b67,b71,b72,b73,b74,b75,b76)
rates=(a1,a2,b12,b21)
```

```python
# model initial conditions
#---------------------------------------------------
OTU1o = OTU1d[0]               
OTU2o = OTU2d[0]                  
#OTU3o = OTU3d[0]
#OTU4o = OTU4d[0]
#OTU5o = OTU5d[0]
#OTU6o = OTU6d[0]
#OTU7o = OTU7d[0]
y0 = [OTU1o, OTU2o]#, OTU3o,OTU4o,OTU5o,OTU6o,OTU7o]      
print OTU1o
print OTU2o
```

```python
# model steps
#---------------------------------------------------
start_time=0.0
end_time=22
intervals=1000
mt=np.linspace(start_time,end_time,intervals)   #evenly spaced numbers over the time space
 
# model index to compare to data
#----------------------------------------------------
findindex=lambda x:np.where(mt>=x)[0][0]
mindex=map(findindex,T)
#=======================================================
```

```python
#3.Score Fit of System
#=========================================================
def score(parms):
    #a.Get Solution to system
    #OTU1,OTU2,OTU3,OTU4,OTU5,OTU6,OTU7,T=eq(rates,y0,start_time,end_time,intervals)    #b.Pick of Model Points to Compare
    OTU1,OTU2,T=eq(rates,y0,start_time,end_time,intervals)
    
    OTU1m=OTU1[mindex]
    OTU2m=OTU2[mindex]
    #OTU3m=OTU3[mindex]
    #OTU4m=OTU4[mindex]
    #OTU5m=OTU5[mindex]
    #OTU6m=OTU6[mindex]
    #OTU7m=OTU7[mindex]
        
    #c.Score Difference between model and data points
    ss=lambda data,model:((data-model)**2).sum()
   
    totalss = ss(OTU1d,OTU1m) #+ ss(OTU2d,OTU2m) #+ ss(OTU3d,OTU3m) + ss(OTU4d,OTU4m) + ss(OTU5d,OTU5m) + ss(OTU6d,OTU6m) + ss(OTU7d,OTU7m)
    
    return totalss
#========================================================
```

```python
#4.Optimize Fit
#=======================================================
fit_score=score(rates)
answ=fmin(score,(rates),full_output=1,maxiter=1000000)
bestrates=answ[0]
bestscore=answ[1]
#K,a1,a2,a3,a4,a5,a6,a7,b12,b13,b14,b15,b16,b17,b21,b23,b24,b25,b26,b27,b31,b32,b34,b35,b36,b36,b37,b41,b42,b43,b45,b46,b47,b51,b52,b53,b54,b56,b57,b61,b62,b63,b64,b65,b67,b71,b72,b73,b74,b75,b76=answ[0]
#newrates=(K,a1,a2,a3,a4,a5,a6,a7,b12,b13,b14,b15,b16,b17,b21,b23,b24,b25,b26,b27,b31,b32,b34,b35,b36,b36,b37,b41,b42,b43,b45,b46,b47,b51,b52,b53,b54,b56,b57,b61,b62,b63,b64,b65,b67,b71,b72,b73,b74,b75,b76)
a1,a2,b12,b21=answ[0]
newrates=(a1,a2,b12,b21)
print newrates
#=======================================================
```

```python
#5.Generate Solution to System
#=======================================================
#OTU1,OTU2,OTU3,OTU4,OTU5,OTU6,OTU7,T=eq(newrates,y0,start_time,end_time,intervals)
OTU1,OTU2,T=eq(newrates,y0,start_time,end_time,intervals)
```

```python
#6. Plot Solution to System
#=========================================================
plt.figure()
plt.plot(T,OTU1,T,OTU2)#,T,OTU3,T,OTU4,T,OTU5,T,OTU6,T,OTU7)
plt.xlabel('days')
plt.ylabel('relative abundance')
title='Pre C. difficile challenge'
plt.title(title)
plt.show()
#=========================================================
```
