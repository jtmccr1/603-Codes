```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy import integrate
```

```python
#The Model
#=======================================================
def eq(par,initial_cond,start_t,end_t,incr):
    #-time-grid-----------------------------------
    t  = np.linspace(start_t, end_t,incr)
    #differential-eq-system---------------------
    def funct(y,t):
        OTU1i = y[0]
        OTU2i = y[1]
        OTU3i = y[2]
        OTU4i = y[3]
        OTU5i = y[4]
        OTU6i = y[5]
        OTU7i = y[6] 

        #Parameters
        K,a1,a2,a3,a4,a5,a6,a7,b12,b13,b14,b15,b16,b17,b21,b23,b24,b25,b26,b27,b31,b32,b34,b35,b36,b36,b37,b41,b42,b43,b45,b46,b47,b51,b52,b53,b54,b56,b57,b61,b62,b63,b64,b65,b67,b71,b72,b73,b74,b75,b76 = par 


        OTU1 = a1*OTU1i*(1-(OTU1i/K) + OTU1i*(b12*OTU2i+b13*OTU3i+b14*OTU4i+b15*OTU5i+b16*OTU6i+b17*OTU7i))
        OTU2 = a2*OTU2i*(1-(OTU2i/K) + OTU2i*(b21*OTU1i+b23*OTU3i+b24*OTU4i+b25*OTU5i+b26*OTU6i+b27*OTU7i))
        OTU3 = a3*OTU3i*(1-(OTU3i/K) + OTU3i*(b31*OTU1i+b32*OTU2i+b34*OTU4i+b35*OTU5i+b36*OTU6i+b37*OTU7i))
        OTU4 = a4*OTU4i*(1-(OTU4i/K) + OTU4i*(b41*OTU1i+b42*OTU2i+b43*OTU3i+b45*OTU5i+b46*OTU6i+b47*OTU7i))
        OTU5 = a5*OTU5i*(1-(OTU5i/K) + OTU5i*(b51*OTU1i+b52*OTU2i+b53*OTU3i+b54*OTU4i+b56*OTU6i+b57*OTU7i))
        OTU6 = a6*OTU6i*(1-(OTU6i/K) + OTU6i*(b61*OTU1i+b62*OTU2i+b63*OTU3i+b64*OTU4i+b65*OTU5i+b67*OTU7i))
        OTU7 = a7*OTU7i*(1-(OTU7i/K) + OTU7i*(b71*OTU1i+b72*OTU2i+b73*OTU3i+b74*OTU4i+b75*OTU5i+b76*OTU6i))
        
        
        return [OTU1,OTU2,OTU3,OTU4,OTU5,OTU6,OTU7]
        
        
   #integrate------------------------------------
    ds = integrate.odeint(funct,initial_cond,t)
    return (ds[:,0],ds[:,1],ds[:,2],ds[:,3],ds[:,4],ds[:,5],ds[:,6],t)
#=======================================================
        
        
        
        
        
        
```

```python
# model initial conditions
#---------------------------------------------------
OTU1o = 0               
OTU2o = 0                  
OTU3o = 0
OTU4o = 0
OTU5o = 0
OTU6o = 0
OTU7o = 0
y0 = [OTU1o, OTU2o, OTU3o,OTU4o,OTU5o,OTU6o,OTU7o]      
 
```

```python
# model steps
#---------------------------------------------------
start_time=0.0
end_time=15.0
intervals=1000
mt=np.linspace(start_time,end_time,intervals)   #evenly spaced numbers over the time space
 
```

```python
#5.Generate Solution to System
#=======================================================
OTU1,OTU2,OTU3,OTU4,OTU5,OTU6,OTU7,T=eq(rates,y0,start_time,end_time,intervals)
```

```python
def eq(par,initial_cond,start_t,end_t,incr):
     #-time-grid-----------------------------------
     t  = np.linspace(start_t, end_t,incr)
     #differential-eq-system----------------------
     def funct(y,t):
        Si=y[0]
        Zi=y[1]
        Ri=y[2]
        P,d,B,G,A=par
        # the model equations (see Munz et al. 2009)
        f0 = P - B*Si*Zi - d*Si
        f1 = B*Si*Zi + G*Ri - A*Si*Zi
        f2 = d*Si + A*Si*Zi - G*Ri
        return [f0, f1, f2]
     #integrate------------------------------------
     ds = integrate.odeint(funct,initial_cond,t)
     return (ds[:,0],ds[:,1],ds[:,2],t)
#=======================================================
```
