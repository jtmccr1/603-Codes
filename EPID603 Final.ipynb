# Initial Conditions #
The initial conditions are set as the first term in the data array imported from R.  Technically this is day 0, everything that made it through the gauvaging step. In determining the initial starting population sizes I am making a range of assumptions which I hope will be valid enough for the time being.  The first is that by the second time point the most abundant bacteriodetes sp. is at 90% its maximum population level.  I also assume growth rate here is 0.43 as estimated in Marino et al 2014.  And that the bactera were growth exponetially between the time points.  I can then solve the equation $ x_t = x_o(1+0.43)^t $ for $x_0$ where t = 1.  I can then use the percent data to calculate the total inititial population

```python
# model initial conditions
#---------------------------------------------------
x1 = 0.5 * 10**10
x0 = x1/1.43
total_loado = 10**9#x0/OTU1d[0]
print(total_loado)
OTU1o = OTU1d[0]*total_loado                
OTU2o = OTU2d[0]*total_loado             
OTU3o = OTU3d[0]*total_loado  
OTU4o = OTU4d[0]*total_loado  
OTU5o = OTU5d[0]*total_loado  
othero = otherd[0]*total_loado



y0 = [OTU1o, OTU2o,OTU3o,OTU4o,OTU5o,total_loado,othero]    
```

# Setting up the equations #
The model is saved as a function eq which intakes the parameters, initial conditions, the start time, end time and the number of increments over that time span.  It house the group of equations that make up the heart of the ode and the integrator.

$ \Delta \text{OTU}_i = \alpha_i \text{OTU}_i (1-\frac{\text{OTU}-i}{k_i})+\text{OTU}_i(\sum{\beta_{i,j} \text{OTU}_j})  $

where OTU$_j$ is 

$ \text{OTU}_j  = \%\text{OTU}_j \cdot \text{total_load}$ 

and $\%\text{OTU}_j $ is given by the forcing function integrating the data (at least for now)

```python
#The Model
#=======================================================
def eq(par,initial_cond,start_t,end_t,incr,data,Td):
#--------------time-grid-------------------------------------------------------------------
    t  = np.linspace(start_t, end_t,incr)
    #f = interp1d(Td,data[0][0:11])
    

    #------------differential-eq-system--------------------------------------------------------
    def funct(y,t):
        OTU1i = y[0]
        OTU2i = y[1]
        OTU3i = y[2]
        OTU4i = y[3]
        OTU5i = y[4]
        total_loadi=y[5]
        otheri = y[6]
        
        
        
         
        #Parameters
        k1,k2,k3,k4,k5,ko,a1,a2,a3,a4,a5,ao,b12,b13,b14,b15,b1o,b21,b23,b24,b25,b2o,b31,b32,b34,b35,b3o,b41,b42,b43,b45,b4o,b51,b52,b53,b54,b5o,bo1,bo2,bo3,bo4,bo5 = par 

        
        OTU1c = a1*OTU1i*(1-(OTU1i/k1)) + OTU1i*(b12*OTU2i+b13*OTU3i+b14*OTU4i+b15*OTU5i+b1o*otheri)
        OTU2c = a2*OTU2i*(1-(OTU2i/k2)) + OTU2i*(b21*OTU1i+b23*OTU3i+b24*OTU4i+b25*OTU5i+b2o*otheri)
        OTU3c = a3*OTU3i*(1-(OTU3i/k3)) + OTU3i*(b31*OTU1i+b32*OTU2i+b34*OTU4i+b35*OTU5i+b3o*otheri)
        OTU4c = a4*OTU4i*(1-(OTU4i/k4))+ OTU4i*(b41*OTU1i+b42*OTU2i+b43*OTU3i+b45*OTU5i+b4o*otheri)
        OTU5c = a5*OTU5i*(1-(OTU5i/k5)) + OTU5i*(b51*OTU1i+b52*OTU2i+b53*OTU3i+b54*OTU5i+b5o*otheri)
        otherc = ao*otheri*(1-(otheri/ko)) + otheri*(bo1*OTU1i+bo2*OTU2i+bo3*OTU3i+bo4*OTU4i+bo5*OTU5i)
        
        total_loadc = OTU1c + OTU2c + OTU3c + OTU4c + OTU5c + otherc  # changes in the total bacterial load only equal the sum of the changes of each OTU
        
        return [OTU1c,OTU2c,OTU3c,OTU4c,OTU5c,total_loadc,otherc]
        
        
   #integrate------------------------------------
    ds = integrate.odeint(funct,initial_cond,t)
    return (ds[:,0],ds[:,1],ds[:,2],ds[:,3],ds[:,4],ds[:,5],ds[:,6],t)
#=======================================================
        
        
```

```python
# model steps
#---------------------------------------------------
start_time=0.0
end_time=22
intervals=1000
mt=np.linspace(start_time,end_time,intervals)   #evenly spaced numbers over the time space
 
# model index to compare to data
#----------------------------------------------------
findindex=lambda x:np.where(mt>=x)[0][0]
mindex=map(findindex,Td)
#=======================================================
```

```python
#3.Score Fit of System
#=========================================================
def score(parms):
    #a.Get Solution to system
   #b.Pick of Model Points to Compare
    OTU1,OTU2,OTU3,OTU4,OTU5,total_load,other,T=eq(parms,y0,start_time,end_time,intervals)
    
    OTU1m=OTU1[mindex]/total_load[mindex]  #Divide by the total bacterial population to get relative abundance
    OTU2m=OTU2[mindex]/total_load[mindex]
    OTU3m=OTU3[mindex]/total_load[mindex]
    OTU4m=OTU4[mindex]/total_load[mindex]
    OTU5m=OTU5[mindex]/total_load[mindex]
    otherm=other[mindex]/total_load[mindex]
    
        
    #c.Score Difference between model and data points
    ss=lambda data,model:sum(((data-model)**2))
   
    #totalss = ss(OTU1d,OTU1m)/np.mean(OTU1m) + ss(OTU2d,OTU2m)/np.mean(OTU2m) + ss(OTU3d,OTU3m)/np.mean(OTU3m) #+ ss(OTU4d,OTU4m)/np.mean(OTU4m) + ss(OTU5d,OTU5m)/np.mean(OTU5m) + ss(OTU6d,OTU6m)/np.mean(OTU6m) + ss(OTU7d,OTU7m)/np.mean(OTU7m)
    totalss = ss(OTU1d,OTU1m) + ss(OTU2d,OTU2m) + ss(OTU3d,OTU3m) + ss(OTU4d,OTU4m) + ss(OTU5d,OTU5m)+ss(otherd,otherm) 
    
    return totalss
#========================================================
```

```python
#4.Optimize Fit
#=======================================================
answ=fmin(score,(rates),full_output=1,maxfun=1000000, maxiter = 1000000)
bestrates=answ[0]
bestscore=answ[1]
k1,k2,k3,k4,k5,ko,a1,a2,a3,a4,a5,ao,b12,b13,b14,b15,b1o,b21,b23,b24,b25,b2o,b31,b32,b34,b35,b3o,b41,b42,b43,b45,b4o,b51,b52,b53,b54,b5o,bo1,bo2,bo3,bo4,bo5=answ[0]
newrates=(k1,k2,k3,k4,k5,ko,a1,a2,a3,a4,a5,ao,b12,b13,b14,b15,b1o,b21,b23,b24,b25,b2o,b31,b32,b34,b35,b3o,b41,b42,b43,b45,b4o,b51,b52,b53,b54,b5o,bo1,bo2,bo3,bo4,bo5)
print newrates
#=======================================================
```
